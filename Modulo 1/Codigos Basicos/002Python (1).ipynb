# funçao print ()
A funcao print () e uma das mais importantes e usadas na linguagem Python. sua funcao basicamente, exiber mensagens na tela ou envia-las para outro dispositivo, como imprimir dentro de um arquivo de texto.

No python 3, print() e uma funcao interna, de um modo que nao e necessario importar nenhuma biblioteca para poder utiliza-la. basta chama-la e passar os argumentos necessarios

```python
print("meu primeiro print!") # todo texto deve estar em aspas simples ou duplas
```

```python
print('meu segundo print')
```

```python
print("d'agua") # quando a palavra contem aspas inicias deve ser diferentes
```

```python
print(20 + 15) # o Python realiza a operaçao que esta dentro do parenteses
```

```python
print("20+15") # a operaçao entre aspas sera apenas impressa nao executada, pois o python entende isso como textos
```

```python
print("20+15", 20 + 15)
```

```python
n1 = "1"
n2 = "3"
print(n1 + n2) # neste exemplo o Python faz a  concatenaçao das 2 strings
```

```python
nome = "diogo"
sobrenome = "trelha"
nome_completo = nome + " " + sobrenome
print(nome_completo)
```
